```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.pyplot as plt
data = []
with open("pdur.txt","r") as f:
    for i in f:
        items=i.split()
        data.append( (float(items[1]))  )
data = np.array(data)
plt.hist(data, bins=100, facecolor="blue", edgecolor="black", alpha=0.7)
plt.xlabel("phone dur")
plt.ylabel("frequency")
plt.title("historgram")
plt.show()
```

```python
logdata = np.log(data)
plt.hist(logdata, bins=100, facecolor="blue", edgecolor="black", alpha=0.7)
plt.xlabel("log phone dur")
plt.ylabel("frequency")
plt.title("historgram")
plt.show()
```
